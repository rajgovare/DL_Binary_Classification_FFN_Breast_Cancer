```python
# Import the required libraries
import keras
import tensorflow
import pandas as pd
```

```python
# Load the breast cancer dataset from scikit-learn
from sklearn.datasets import load_breast_cancer
dataset=load_breast_cancer()
```

```python
# Extract the features (X) and labels (y) from the dataset
x=dataset.data
x.shape
```

```python
y=dataset.target
y.shape
```

```python
# Split the dataset into training and testing sets
from sklearn.model_selection import train_test_split
```

```python
xtrain,xtest,ytrain,ytest=train_test_split(x,y,test_size=0.25,random_state=1)
xtrain.shape
```

```python
xtest.shape
```

```python
ytrain.shape
```

```python
# Build a Feed Forward Neural Network (FFN) model using Keras
from keras.models import Sequential
from keras.layers import Dense

# Initialize the sequential model
nnmodel=Sequential()
```

```python
# Add an input layer with 18 neurons and ReLU activation function
# The input layer takes 30 features from the dataset
nnmodel.add(Dense(18,activation='relu',input_dim=30))

# Add a hidden layer with 12 neurons and ReLU activation function
nnmodel.add(Dense(12,activation='relu'))


# Add the output layer with 1 neuron and sigmoid activation function
# The output layer predicts the class (binary: 0 or 1)
nnmodel.add(Dense(1,activation='sigmoid'))
```

```python
# Compile the model
nnmodel.compile(loss='binary_crossentropy',optimizer='adam',metrics=['accuracy'])
```

```python
# Train the model using the training data
nnmodel.fit(xtrain,ytrain,epochs=5000,batch_size=40)
```

```python
ytest
```

```python
# Make predictions using the trained model on the test data
predictions=nnmodel.predict(xtest)
predictions
```

```python
predictions.size
```

```python
# Convert the continuous predictions (probabilities) into binary class labels (0 or 1)
class_labels=[]
predictions.size
for i in range(143):
    if(predictions[i]>=0.5):
        class_labels.append(1)
    else:
        class_labels.append(0)
```

```python
# Print the final predicted class labels
class_labels
```

```python
# Import the accuracy_score and confusion_matrix from sklearn to evaluate model performance
from sklearn.metrics import accuracy_score,confusion_matrix

# Calculate the accuracy score by comparing the predicted labels with the actual test labels
accuracy_score(ytest,class_labels)
```

```python
# Generate the confusion matrix to show the model's performance
confusion_matrix(ytest,class_labels)
```
